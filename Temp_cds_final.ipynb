```python
import pandas as pd
import numpy as np
import datetime
from datetime import timedelta, date
#import datetime
from datetime import datetime
import time
```

```python
year = str(2006)
```

```python
csv_input = pd.read_csv("temp_cen_"+str(year)+"_data.csv", encoding= 'unicode_escape', low_memory = False)
```

```python
csv_input
```

```python
[rows, cols] = csv_input.shape
rows
```

```python
csv_input.iloc[:,0]
```

```python
# define the date of 1 April and 31 october
# start column index of 1 April
month = 4
day = 1
april_1 =  datetime(int(year), month, day)
start_day = int(april_1.strftime('%j'))

# end index of column is 31 october so, 1 Nov end day 
month = 11
day = 1
nov_1 =   datetime(int(year), month, day)
end_day = int(nov_1.strftime('%j'))

print(start_day, end_day)
```

```python
csv_input.iloc[0, start_day:end_day]  # *****make sure in leap years case this should start from 92************
```

```python
start = time.time()

data = []
year = str(year)
for i in range(rows):
    grid_id = csv_input.iat[i, 0] 
    for j in range(start_day,end_day):
        
        # convert day to datetime
        day_value = str(j)
        day_value.rjust(3 + len(day_value), '0')
        date_value = datetime.strptime(year + "-" + day_value, "%Y-%j")
        
        temp_value = csv_input.iat[i,j]
        item = [grid_id, day_value, temp_value]
        data.append([grid_id, date_value, temp_value])
        
end = time.time()
print ("Time elapsed:", end - start)
```

```python
out_df = pd.DataFrame(data, columns = ['Grid_id', 'date', 'T'])
```

```python
start = time.time()
out_df.to_csv("Temp_"+str(year)+".csv", index=False)
end = time.time()
print ("Time elapsed:", end - start)
```
