### Extract Unique Townships to csv

```python
print("num of Unique Townships", csv_input['twpid'].nunique())
unique_townships = csv_input['twpid'].unique()
print("unique townships", unique_townships)
```

```python
unique_townships_df = pd.DataFrame(unique_townships, columns = ['twpid'])
print(unique_townships_df)
unique_townships_df.to_csv('unique_townships.csv', index = False)
```

### Extract the unique Townships locations to csv

```python
# read the Township metdata file

township_metadata = pd.read_csv("township lat long eastings elevation.csv", encoding= 'unicode_escape', low_memory = False)
township_metadata = township_metadata.rename(columns = {'Twpid':'twpid'})
print( township_metadata)
```

```python
unique_townships_metadata = pd.merge(unique_townships_df, township_metadata, 
                                     on = 'twpid', 
                                     how = 'inner')
print(unique_townships_metadata)

unique_townships_metadata.to_csv('unique_townships_metadata.csv', index = False)
```

```python
unique_townships_metadata = pd.merge(unique_townships_df, township_metadata, 
                                     on = 'twpid', 
                                     how = 'inner')
print(unique_townships_metadata)
```
