```python
import pandas as pd
import numpy as np
import datetime
from datetime import timedelta, date

from datetime import datetime
```

```python
year = str(2001)
```

```python
# Import Data
shape = pd.read_csv('Joined_fire_shapes_new.csv', encoding= 'unicode_escape')
shape_new = shape.dropna(subset=['id_2'])
shape_new = shape_new.drop(columns = ['id_2'])
shape_new.rename(columns = {'id': 'Grid_id'}, inplace = True)
```

```python
custom_date_parser = lambda x: datetime.strptime(x, "%Y-%m-%d")
temp = pd.read_csv('Temp_'+year+'.csv',
                 parse_dates=['date'],
                date_parser=custom_date_parser)
```

```python
## Merge two dataframes
temp = pd.merge(temp, shape_new, 
                    on = ['Grid_id'], 
                    how =  'inner')
```

```python
met_ndvi_temp = pd.merge(met_ndvi, temp, 
                    on = ['Grid_id', 'date'], 
                    how =  'inner')
```

```python
## Read topographical variables

slope_elev_asp = pd.read_csv('slope_elev_asp_output.csv')
print(slope_elev_asp.head())
print(slope_elev_asp.shape)
```

```python
data = pd.merge(met_ndvi_temp, slope_elev_asp, 
                      on = 'Grid_id', 
                    how =  'inner')
```

```python
print(data.head())
print(data.shape)
```
