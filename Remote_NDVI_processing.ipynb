```python
import pandas as pd
import numpy as np
import datetime
from datetime import timedelta, date
```

```python
year = 2009
```

```python
csv_input = pd.read_csv("Remote_NDVI_"+str(year)+".csv", encoding= 'unicode_escape')
```

```python
csv_input = pd.read_csv("Remote_NDVI_"+str(year)+".csv", encoding= 'unicode_escape')
copy = csv_input[['id','_b1_mean']].copy(deep = False)
month = 3
day = 31
date =  datetime.datetime(year, month, day)
april_1_5 = []
for i in range(5):
    copy = csv_input[['id','_b1_mean']].copy(deep = False)
    copy.rename(columns = {'id': 'Grid_id', '_b1_mean': 'NDVI' } , inplace=True)
    date =  (date+ timedelta(days=1))
    copy['date'] = date
    april_1_5.append(copy)   
month = 4
day = 6
total_weeks = 26
date =  datetime.datetime(year, month, day-1)
temp = []
for i in range(1, total_weeks+1):
    for j in range(7):
        copy = csv_input[['id','_b'+ str(i)+'_mean']].copy(deep = False)
        copy.rename(columns = {'id': 'Grid_id', '_b'+ str(i)+'_mean': 'NDVI' } , inplace=True)
        date =  (date+ timedelta(days=1))
        copy['date'] = date
        temp.append(copy)
month = 10
day = 4
date =  datetime.datetime(year, month, day)
oct_5_31 = []
for i in range(27):
    copy = csv_input[['id','_b26_mean']].copy(deep = False)
    copy.rename(columns = {'id': 'Grid_id', '_b26_mean': 'NDVI' } , inplace=True)
    date =  (date+ timedelta(days=1))
    copy['date'] = date
    oct_5_31.append(copy)      
april_start = pd.concat(april_1_5, axis = 0, ignore_index = True)
april_oct = pd.concat(temp, axis=0, ignore_index=True)
oct_end = pd.concat(oct_5_31, axis = 0, ignore_index = True)
output= pd.concat([april_start, april_oct, oct_end], axis=0, ignore_index=True)
output.to_csv('NDVI_'+str(year)+'.csv', index=False)
```
